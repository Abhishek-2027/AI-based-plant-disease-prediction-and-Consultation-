```python
import cv2
```

```python
image_path="/kaggle/input/new-plant-diseases-dataset/test/test/TomatoHealthy3.JPG"
```

```python
img=cv2.imread(image_path)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("image of testing leaf")
plt.show()
```

```python

import numpy as np
img = image.load_img(image_path, target_size=(128,128))
img_arr = image.img_to_array(img)
img_arr = np.expand_dims(img_arr, axis=0)   # add batch dimension if needed
print(img_arr.shape)
```

```python
pred=model.predict(img_arr)
pred.shape,pred
```

```python
pred_index=np.argmax(pred)
pred_index
```

```python
valid=tf.keras.utils.image_dataset_from_directory(
     '/kaggle/input/new-plant-diseases-dataset/New Plant Diseases Dataset(Augmented)/New Plant Diseases Dataset(Augmented)/valid/',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=False,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    
)
```

```python
class_name=valid.class_names
class_name
```

```python
pred_class=class_name[pred_index]
print("disease name is : ",pred_class)
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.title("image of prediction leaf")
plt.show()
```
